```python
import requests
from bs4 import BeautifulSoup
import json


url = "https://www.amazon.fr/gp/bestsellers/?ref_=nav_em_cs_bestsellers_0_1_1_2"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/144.0.0.0 Safari/537.36 Edg/144.0.0.0",
    "Accept-Language": "fr-FR,fr;q=0.9"
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")
amazone="https://www.amazon.fr"
print(soup.title)
print(response.status_code)


#<a aria-label=" Les meilleures ventes en Mode - Voir plus" class="a-link-normal" href="/gp/bestsellers/fashion/ref=zg_bs_fashion_sm">Voir plus</a>
links = {
    a.get("aria-label").strip(): amazone + a.get("href")
    for a in soup.find_all("a", class_="a-link-normal")
    if a.get("aria-label") and a.get("href") and "/gp/bestsellers/" in a.get("href")
}

print("Nombre de catégories :", len(links))

stock=[]

for category,link in links.items():
    izem=link
    while izem :
       response = requests.get(izem, headers=headers)
       soup=BeautifulSoup(response.text, "html.parser")
       uccanen=soup.select("div.p13n-sc-uncoverable-faceout")
       print("Produits trouvés :", len(uccanen))
       for uccen in uccanen :
          nom_tag = uccen.select_one("div._cDEzb_p13n-sc-css-line-clamp-3_g3dy1")
          note_tag = uccen.select_one("span.a-icon-alt")
          prix_tag = uccen.select_one("span.p13n-sc-price")
          # Vérification que le prix a été trouvé
          if prix_tag:
                prix = prix_tag.get_text(strip=True)
          else:
                prix_tag = uccen.select_one("span._cDEzb_p13n-sc-price_3mJ9Z")
                if prix_tag:
                    prix = prix_tag.get_text(strip=True)
                else:
                    prix = None  # Si aucun prix n'est trouvé, on assigne None
          
          produt = nom_tag.get_text(strip=True) if nom_tag else None
          note = note_tag.get_text(strip=True) if note_tag else None
          

          print("DEBUG →", produt, note, prix)
          stock.append({
              "Category":category,
              "Product_name":produt,
              "Note":note,
              "Prix":prix,
          })
       izirdi=soup.select("li.a-last a")
       if izirdi:
         izem=amazone+izirdi[0].get('href')
       else :
         izem=None



with open('file.json', 'w', encoding='utf-8') as json_file:
    json.dump(stock, json_file, ensure_ascii=False, indent=4)

print(" file.json est créé")
```
