```python
import requests
from bs4 import BeautifulSoup
import csv
import os

date = input("please enter a date in the following format DD/MM/YYYY : ")

page = requests.get(f"https://www.yallakora.com/match-center?date={date}")

def main(page):
    src = page.content
    soup = BeautifulSoup(src, "lxml")
    match_details = []

    championships = soup.find_all("div", class_="matchesList")

    for championship in championships:
        title = championship.find("h2")
        if not title:
            continue

        championship_title = title.text.strip()

        matches = championship.find_all("div", class_="liItem")

        for match in matches:
            team_A = match.find("div", class_="teamA").p.text.strip()
            team_B = match.find("div", class_="teamB").p.text.strip()

            time = match.find("span", class_="time").text.strip()

            score_spans = match.find_all("span", class_="score")
            if len(score_spans) == 2:
               score_text1 = score_spans[0].text.strip()
               score_text2 = score_spans[1].text.strip()
    
    # Handle cases like "2", "0", or part of "2-0"
               if score_text1.isdigit() and score_text2.isdigit():
                  score1 = int(score_text1)
                  score2 = int(score_text2)
                  #score = score1 "-" score2
                  score = f"{int(score1)}-{int(score2)}"
               else:
                  score = "N/A"  # Dash or other non-numeric
            else:
                 score = "N/A"

            match_details.append({
                "Date": date,
                "Championnat": championship_title,
                "Equipe A": team_A,
                "Equipe B": team_B,
                "Heure": time,
                "Score": score
            })

    if not match_details:
        print("❌ Aucun match trouvé")
        return

    # création du fichier CSV
    with open("match.csv", "w", newline="", encoding="utf-8-sig") as file:
        
        writer = csv.DictWriter(
        file,
        fieldnames=match_details[0].keys(),
        delimiter=";"
    )
        writer.writeheader()
        writer.writerows(match_details)

    

        

    print("✅ Fichier match.csv créé avec succès")

main(page)
```
